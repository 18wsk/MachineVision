```python
import cv2 
from matplotlib import pyplot as plt
import numpy as np
import math # need math.sqrt() for Euclidean Alg
LARGE_NUMBER = int(2021202020192018)

def numberOfClusters(k):
    pass
    
def getTrackBarPosition():
    return int(cv2.getTrackbarPos('K', "K-MEANS")) # returns value of "Intensity" based on trackbar position

def randomCenters(number_of_clusters):
    list_of_centers.clear()
    center_BGR.clear()
    if number_of_clusters != 0:
        for i in range(number_of_clusters):
            row = np.random.randint(0, img_height) # number of random pixels generated based off K from trackbar --> threshold based off image dimensions
            col = np.random.randint(0, img_width)
            list_of_centers.append((col, row))
            center_BGR.append(baboon[col, row])
            # print("Center {} is {}".format(i, list_of_centers[i]))
            
def VoronoiRegion(k_val): # comparing seed_pixel and current_Pixel being evaluated to see if its RGB values are within intensity threshold
    list_of_regions.clear()
    for col in range(img_width):
        for row in range(img_height):
            min_distance = LARGE_NUMBER # random large number to ensure first delta_E is always first min_distance --> then the others are compared against first delta_E after
            min_center = 0
            for i in range(k_val):
                R = (int(center_BGR[i][2]) - int(baboon[col,row][2]))**2 # BGR image
                G = (int(center_BGR[i][1]) - int(baboon[col,row][1]))**2
                B = (int(center_BGR[i][0]) - int(baboon[col,row][0]))**2 # using exact same euclidean layout from my pre-lab 2 --> distance calculated from random center BGR - every pixel BGR
                delta_E = math.sqrt(R+G+B)
                
                if delta_E <= min_distance:
                    min_distance = delta_E # if the min distance of this pixel compaired to current evaluating center --> this min center becomes its new one
                    min_center = i
    
            list_of_regions.append([cpy[col,row], (col,row), min_distance, min_center]) # list of regions is a nested list of information for each o the pixels
    return list_of_regions    
    

def KMeans(k_val):
    sum_R_values.clear()
    sum_G_values.clear()
    sum_B_values.clear()
    current_region = VoronoiRegion(k_val) # current region now holds list of current regions and their nested info
    for i in range(len(list_of_centers)):
        sum_R_values.append([]) # creating nested arrays for each of the center values --> easier to iterate through one main list
        sum_G_values.append([])
        sum_B_values.append([])
        for j in range(len(current_region)):
           #  print("Region min center {} vs center list {}".format(current_region[j][3], i))
            if current_region[j][3] == i: # if the pixel of this region has a min_center matching one of min centers from min_center list --> add it to that min centers RGB sum lists
                sum_R_values[i].append(current_region[j][0][2]) # accessing RGB values of ech region and adding them to corresponding RGB list
                sum_G_values[i].append(current_region[j][0][1])
                sum_B_values[i].append(current_region[j][0][0])
        # print("Number of pixels for center {} is {}".format(i, len(sum_R_values[i])))   
    
    mean_channels.clear()
    for x in range(len(list_of_centers)):
        mean_channels.append([int(np.mean(sum_B_values[x])), int(np.mean(sum_G_values[x])), int(np.mean(sum_R_values[x]))]) # using numpy library built-in mean method to quickly calculate avg of long lists
     
    for m in range(len(current_region)):
        for n in range(len(list_of_centers)): # for everypoint check what center is closest and assign the corresponding mean of that voronoiRegion to its rgb value
            if current_region[m][3] == n:
                cpy[current_region[m][1][0], current_region[m][1][1]][2] = mean_channels[n][2] # now replace every pixel RGB of copy image with that of mean value
                cpy[current_region[m][1][0], current_region[m][1][1]][1] = mean_channels[n][1]
                cpy[current_region[m][1][0], current_region[m][1][1]][0] = mean_channels[n][0] 
    print("DONE") # when DONE is printed in console user will know operation is complete

    
    
baboon = cv2.imread("baboon.jpg")
cpy = baboon.copy()
img_width, img_height, img_channel = baboon.shape # initializing image variables

global list_of_centers, mean_channels, current_k, sum_R_values, sum_G_values, sum_B_values, current_region
current_region = []
list_of_centers = []
list_of_regions = []
mean_channels = []
sum_R_values = []
sum_G_values = []
sum_B_values = []
center_BGR = []
current_k = 1
count = 0
cv2.namedWindow("K-MEANS", cv2.WINDOW_NORMAL)
cv2.createTrackbar('K', "K-MEANS",1, 22, numberOfClusters)# imlementing trackbar function --> this will create a variable intensity threshold to which the homogenity criterion will be based off of

while True:
    cv2.imshow("K-MEANS", cpy)
    k = getTrackBarPosition()
    if k != current_k:
        cpy = baboon.copy() # image is refreshed in backend so KMEANS can be re-applied
        current_k = k
        randomCenters(current_k)
        KMeans(current_k)
        
        for x in range(len(list_of_centers)): # this now checks if center is within reasonable distance from mean --> checking R/G/B differences individually
            if abs(mean_channels[x][0]-baboon[list_of_centers[x]][0]) > 50 or abs(mean_channels[x][0]-baboon[list_of_centers[x]][0]) > 50 or abs(mean_channels[x][0]-baboon[list_of_centers[x]][0]) > 50:
                print("Bad Center", baboon[list_of_centers[x]], " VS ", mean_channels[x]) # if it is not the center value will be printed and replaced with mean RGB value
                center_BGR[x] = mean_channels[x]
                count += 1
            if x == len(list_of_centers)-1 and count > 0: # if any of the centers are bad --> KMEANS is recalculated with new centers that are the mean RGB values
                randomCenters(current_k)
                KMeans(current_k)
                
    
    if cv2.waitKey(1) == 27: # if user hits ESC key --> window is exited
            break 
cv2.destroyAllWindows()
```
