```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import sklearn

print(cv2.__version__)
print(np.__version__)
print(sklearn.__version__)
```

```python
print("Hello World")
img = cv2.imread("lena.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Lena Color")
```

```python
plt.imshow(gray, cmap = 'gray') # changing images colour map
plt.title("Lena Grayscale")

# cmap = The Colormap instance or registered colormap name used to map scalar data to colors. This parameter is ignored for RGB(A) data.
```

```python
print(img.shape) # rows, coloumns, channels
print(img.size) # total number of pixels
print(img.dtype) # images data type
```

```python
# FLIPPING AN IMAGE
img_width, img_height, img_channel = img.shape # initializing array dimensions for looping purposes
hor_flipped_img = np.zeros((img_height,img_width,3), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        for k in range(img_channel):
            hor_flipped_img[i][j][k] = img[i][img_width-j-1][k] # horizontally flipped image has a reveresed order of columns in its array values
        
plt.imshow(cv2.cvtColor(hor_flipped_img, cv2.COLOR_BGR2RGB))        
plt.title("Lena Colour Horizontally Flipped")
```

```python
# FLIPPING AN IMAGE
gray_vert_flipped_img = np.zeros((img_height, img_width, 3), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        for k in range(img_channel):
            gray_vert_flipped_img[i][j][k] = img[img_height-i-1][j][k] # vertically flipped image has a reveresed order of rows in its array values

plt.imshow(cv2.cvtColor(gray_vert_flipped_img, cv2.COLOR_BGR2RGB))        
plt.title("Lena Colour Vertically Flipped")
```

```python
# FLIPPING AN IMAGE
img_width, img_height = gray.shape # initializing array dimensions for looping purposes --> NO CHANNEL DIMENSION (Just 1 for gray scale)
gray_hor_flipped_img = np.zeros((img_height,img_width,1), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
            gray_hor_flipped_img[i][j] = gray[i][img_width-j-1] # horizontally flipped image has a reveresed order of columns in its array values --> no need to loop through channel
        
plt.imshow(cv2.cvtColor(gray_hor_flipped_img, cv2.COLOR_BGR2RGB))        
plt.title("Lena Horizontally Flipped")
```

```python
# FLIPPING AN IMAGE
gray_vert_flipped_img = np.zeros((img_height, img_width, 1), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        gray_vert_flipped_img[i][j] = gray[img_height-i-1][j] # vertically flipped image has a reveresed order of rows in its array values

plt.imshow(cv2.cvtColor(gray_vert_flipped_img, cv2.COLOR_BGR2RGB))        
plt.title("Lena Colour Vertically Flipped")
```

```python
lena_colour_negated = np.zeros((img_height, img_width, 3), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        for k in range(img_channel):
            lena_colour_negated[i][j][k] = (255 - img[i][j][k]) # negating values in simple terms is switching the end of the 0-255 spectrum and since the type is unsigned -255 = 255, so black becomes white and white becomes black
            
plt.imshow(cv2.cvtColor(lena_colour_negated, cv2.COLOR_BGR2RGB))        
plt.title("Lena Colour Negated")
```

```python
lena_grayscale_negated = np.zeros((img_height, img_width, 1), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        lena_grayscale_negated[i][j]= (255 - gray[i][j]) # negating values in simple terms is switching the end of the 0-255 spectrum and since the type is unsigned -255 = 255, so black becomes white and white becomes black
            
plt.imshow(cv2.cvtColor(lena_grayscale_negated, cv2.COLOR_BGR2RGB))        
plt.title("Lena Grayscale Negated")
```

```python
lena_red_image = np.zeros((img_height, img_width, 3), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        for k in range(img_channel):
            if k == 0: # if the channel is a part of the RED channel (in this case channel 0) -> set its value to the img val --> if not set to 0 so only red values appear
                lena_red_image[i][j][k] = img[i][j][k]
            else:
                lena_red_image[i][j][k] = 0
            
plt.imshow(lena_red_image)        
plt.title("Lena Red Image")
```

```python
lena_green_image = np.zeros((img_height, img_width, 3), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        for k in range(img_channel):
            if k == 1: # if the channel is a part of the GREEN channel (in this case channel 1) -> set its value to the img val --> if not set to 0 so only green values appear
                lena_green_image[i][j][k] = img[i][j][k]
            else:
                lena_green_image[i][j][k] = 0
            
plt.imshow(lena_green_image)        
plt.title("Lena Green Image")
```

```python
lena_blue_image = np.zeros((img_height, img_width, 3), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        for k in range(img_channel):
            if k == 2: # if the channel is a part of the BLUE channel (in this case channel 1) -> set its value to the img val --> if not set to 0 so only blue values appear
                lena_blue_image[i][j][k] = img[i][j][k]
            else:
                lena_blue_image[i][j][k] = 0
            
plt.imshow(lena_blue_image)        
plt.title("Lena Blue Image")
```

```python
lena_merged_image = np.zeros((img_height, img_width, 3), np.uint8) # initializing empty images with same width, height, channel and dtype

for i in range(img_height):
    for j in range(img_width):
        for k in range(img_channel):
            if k == 0: # when looping through the array take the isloated channels and fill in the merged image values when their channel value is called
                lena_merged_image[i][j][k] = lena_red_image[i][j][k]
            elif k == 1:
                lena_merged_image[i][j][k] = lena_green_image[i][j][k]
            elif k == 2:
                lena_merged_image[i][j][k] = lena_blue_image[i][j][k]
plt.imshow(cv2.cvtColor(lena_merged_image, cv2.COLOR_BGR2RGB))        
plt.title("Lena Merged Image")
```
